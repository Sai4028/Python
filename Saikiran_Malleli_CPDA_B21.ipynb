#### There are 8,358 missing values in the zip column. To fill these, existing zip codes found in the dataset for the same merchant_state and merchant_city combinations will be used as a reference. If neither a matching combination is found nor valid values exist for merchant_state or merchant_city (for example, if they are incorrect or marked as "NA"), the missing zip will be filled with "00000".

```python
# Creating a smaller lookup DataFrame with only the relevant columns
zip_lookup_df = transactions_data_df[['merchant_city', 'merchant_state', 'zip']]
```

```python
# Creating a mapping of (merchant_city, merchant_state) ‚Üí zip, Including only rows where ZIP is not null or empty
zip_mapping = dict(zip(
    zip_lookup_df.loc[zip_lookup_df['zip'].notna() & (zip_lookup_df['zip'] != ''), 
                      ['merchant_city', 'merchant_state']].apply(tuple, axis=1),
    zip_lookup_df.loc[zip_lookup_df['zip'].notna() & (zip_lookup_df['zip'] != ''), 'zip']
))
```

```python
# User defined function to fill ZIP values,
# If merchant_city is 'ONLINE', assign a default ZIP as '00000',  
# Otherwise, fetch ZIP from the mapping based on city and state, 
# If no match is found, assign '00000' as default

def zip_code(row):
  if row['merchant_city'] == 'ONLINE':
     return '00000'
  else:
     return zip_mapping.get((row['merchant_city'], row['merchant_state']), '00000')
```

```python
# Applying the function to the DataFrame to fill missing ZIPs
transactions_data_df['zip'] = transactions_data_df.apply(zip_code, axis = 1)
```

```python
transactions_data_df.isnull().sum()
```

#### From above command we can see that all missing values from transactions data frame are handled

### 7. Make sure state codes are always two capital letters (like CA, NY)

```python
transactions_data_df['merchant_state'].unique()
```

```python
def get_state_code(state):               # creating function for below conditions  
    state = str(state).strip().upper()   # remove spaces & make uppercase
    words = state.split()                 # split state into words
    if len(words) == 1:
        return words[0][:2]              # first 2 letters if only one word
    else:
        return words[0][0] + words[1][0] # first letters of first 2 words
```

```python
# applying above created function
transactions_data_df['merchant_state'] = transactions_data_df['merchant_state'].apply(get_state_code)
```

```python
# Confirmation on whether the merchant state values are as per requested condition with two capital letters
transactions_data_df['merchant_state'].unique()
```

#### 8. How do you turn ZIP codes into a proper 5-digit string, keeping leading zeros?

```python
# To check the zip values
transactions_data_df['zip'].unique()
```

```python
# Removing.0 after every zip code with replace function & fill zeros infront to ensure all zipcodes are of 5 digits,if any zipcode is less than 5 digits..
transactions_data_df['zip'] = transactions_data_df['zip'].astype(str).str.replace(r'\.0$', '', regex=True).str.zfill(5)
```
