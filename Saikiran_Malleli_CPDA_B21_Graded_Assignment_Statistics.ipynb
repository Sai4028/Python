## Question- 1 : For the suburb of Altona, it is postulated that a typical property sells for 800,000 dollars. Use the data at hand to test this assumption. Is the typical property price really $800,000 or has it increased? Use a significance level of 5%.

- H0 : Average property sells for 800000 dollars in suburb of Altona
- H1 : Average property sells for more than 800000 dollars in suburb of Altona

```python
pop_mean = 800000
```

```python
altona_prices = property_df[property_df['Suburb'] == 'Altona']['Price']
```

```python
altona_prices
```

```python
t_test, p_value = stats.ttest_1samp(altona_prices, pop_mean)
```

```python
t_test
```

```python
p_value
```

#### Given significance_level(alpha) = 0.05

```python
if p_value <= 0.05:
    print('Reject H0 : Average Property sells more than $800000')
else:
    print('Fail to Reject H0 : Average Property sells at $800000')    
```

### From the T-test, conclusion is failed to reject the null hypothesis, so average property sells at $800000 in the suburb of Altona is correct.

## Question-2 : For the year 2016, is there any difference in the prices of properties sold in the summer months vs winter months? Consider months from October till March as winter months and rest as summer months. Use a significance level of 5%.

- H0 : There is no difference in mean property prices between summer and winter months in 2016.
- H1 : There is difference in mean property prices between summer and winter months in 2016.

```python
property_df['Date'] = pd.to_datetime(property_df['Date'], format='%d/%m/%Y')
```

```python
property_df['Sold_Year'] = property_df['Date'].dt.year # extracting sold_year from date
```

```python
property_df['Season'] = property_df['Date'].dt.month.apply(lambda m: 'Winter' if m >= 10 or m <= 3 else 'Summer') # lambda fun for season using months
```

```python
property_df[['Date', 'Sold_Year', 'Season']].head()
```

```python
summer_prices = property_df[(property_df['Season'] == 'Summer') & (property_df['Sold_Year'] == 2016)]['Price']
```

```python
summer_prices
```

```python
winter_prices = property_df[(property_df['Season'] == 'Winter') & (property_df['Sold_Year'] == 2016)]['Price']
```

```python
winter_prices
```

```python
stats.ttest_ind(summer_prices, winter_prices)
```

```python
ttest,p_value = stats.ttest_ind(summer_prices, winter_prices)
```

```python
ttest
```

```python
p_value
```

```python
if p_value <= 0.05:
    print('Reject H0 : Difference in mean property prices between summer and winter months in 2016')
else:
    print('Fail to Reject H0 : No difference in mean property prices between summer and winter months in 2016')    
```

### From above T-test, the conclusion is, there is difference in mean property prices between summer and winter months in 2016
